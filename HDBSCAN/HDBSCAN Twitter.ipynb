HDBSCAN

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import HDBSCAN
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score, silhouette_score
from tqdm import tqdm
```

```python
# Load the Twitter dataset
twitter = pd.read_csv('twitter.csv')

print(f"Dataset Shape: {twitter.shape}")
print(f"\nColumns: {twitter.columns.tolist()}")
print(f"\n Data Sample")
print(twitter.head())
print(f"\nData Types:")
print(twitter.dtypes)
```

```python
# Convert timestamp to datetime
# Format: YYYYMMDDHHmmss (e.g., 20130112000000 = 2013-01-12 00:00:00)
twitter['datetime'] = pd.to_datetime(twitter['timestamp'], format='%Y%m%d%H%M%S')

# Extract time components
twitter['hour'] = twitter['datetime'].dt.hour
twitter['minute'] = twitter['datetime'].dt.minute
twitter['day_of_week'] = twitter['datetime'].dt.dayofweek  # 0=Monday, 6=Sunday

# Create fractional hour (hour + minute/60)
twitter['hour_frac'] = twitter['hour'] + twitter['minute'] / 60.0

# Cyclical encoding for hourly pattern (24h cycle)
twitter['hour_sin'] = np.sin(2 * np.pi * twitter['hour_frac'] / 24)
twitter['hour_cos'] = np.cos(2 * np.pi * twitter['hour_frac'] / 24)

# Cyclical encoding for weekly pattern (7d cycle)
twitter['day_sin'] = np.sin(2 * np.pi * twitter['day_of_week'] / 7)
twitter['day_cos'] = np.cos(2 * np.pi * twitter['day_of_week'] / 7)

print("✓ Cyclical time features created")
print(f"\nHourly cycle features:")
print(f"  hour_sin range: [{twitter['hour_sin'].min():.3f}, {twitter['hour_sin'].max():.3f}]")
print(f"  hour_cos range: [{twitter['hour_cos'].min():.3f}, {twitter['hour_cos'].max():.3f}]")
print(f"\nWeekly cycle features:")
print(f"  day_sin range: [{twitter['day_sin'].min():.3f}, {twitter['day_sin'].max():.3f}]")
print(f"  day_cos range: [{twitter['day_cos'].min():.3f}, {twitter['day_cos'].max():.3f}]")
```

```python
# Visualize cyclical encoding
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Hourly cycle
axes[0].scatter(twitter['hour_sin'][:1000], twitter['hour_cos'][:1000], 
                c=twitter['hour'][:1000], cmap='twilight', alpha=0.6, s=20)
axes[0].set_xlabel('hour_sin')
axes[0].set_ylabel('hour_cos')
axes[0].set_title('Hourly Cycle: Midnight and 23:59 are close')
axes[0].set_aspect('equal')
circle1 = plt.Circle((0, 0), 1, fill=False, linestyle='--', color='gray')
axes[0].add_patch(circle1)

# Weekly cycle
axes[1].scatter(twitter['day_sin'][:1000], twitter['day_cos'][:1000], 
                c=twitter['day_of_week'][:1000], cmap='Set2', alpha=0.6, s=20)
axes[1].set_xlabel('day_sin')
axes[1].set_ylabel('day_cos')
axes[1].set_title('Weekly Cycle: Sunday and Monday are close')
axes[1].set_aspect('equal')
circle2 = plt.Circle((0, 0), 1, fill=False, linestyle='--', color='gray')
axes[1].add_patch(circle2)

plt.tight_layout()
plt.show()
```

```python
twitter['lat_rad'] = np.radians(twitter['latitude'])
twitter['long_rad'] = np.radians(twitter['longitude'])

print("✓ Geographic features created")
print(f"\nRadian coordinates:")
print(f"  lat_rad range: [{twitter['lat_rad'].min():.4f}, {twitter['lat_rad'].max():.4f}]")
print(f"  long_rad range: [{twitter['long_rad'].min():.4f}, {twitter['long_rad'].max():.4f}]")
```

```python
# Feature Set 1: Radians (for Haversine) + Time
features_spatial = twitter[['lat_rad', 'long_rad', 'hour_sin', 'hour_cos']].copy()
# Feature Set 2: Radians (for Haversine) + Hourly + Weekly cycles
features_spatial_full = twitter[['lat_rad', 'long_rad', 'hour_sin', 'hour_cos', 'day_sin', 'day_cos']].copy()
print("✓ Feature sets created")
print(f"\nFeature Set 1 (Spatial + Hourly): {features_spatial.shape}")
print(features_spatial.columns.tolist())
print(f"\nFeature Set 2 (Spatial + Full Time): {features_spatial_full.shape}")
print(features_spatial_full.columns.tolist())

# Apply MinMaxScaler to all features
scaler1 = MinMaxScaler()
features_set1_scaled = scaler1.fit_transform(features_spatial)
scaler2 = MinMaxScaler()
features_set2_scaled = scaler2.fit_transform(features_spatial_full)

min_sample_size = [10, 50, 100, 500]
min_cluster_size = [50, 100, 200, 500]
print(f"Grid search parameters:")
print(f"\nMinimum Cluster Size: {min_cluster_size}")

# Sample data for faster tuning (optional)
SAMPLE_SIZE = 25000
#sample_indices = np.arange(SAMPLE_SIZE)  
idx = np.random.choice(len(twitter), size=SAMPLE_SIZE, replace=False) # Random 25000 indices

sample_set1 = features_set1_scaled[idx]            # shape: (n, 4)
sample_set2 = features_set2_scaled[idx]

labels_sample = twitter['timezone'].iloc[idx].to_numpy()  # Fixed: use indices, not array

print(f"\n✓ Sample created: {len(sample_set1)} points")
print(f"  Features shape: {sample_set1.shape}")
print(f"  Labels shape: {labels_sample.shape}")
```

Parameter Testing for Feature Set 1

```python
results = []
total = len(min_sample_size) * len(min_cluster_size)
count = 1

print("\nStarting grid search for HDBSCAN...\n")

for idx, sample_size in enumerate(min_sample_size,1):
    for idx2, min_size in enumerate(min_cluster_size, 1):
        print(f"\n[{count}/{total}] Testing min_sample_size={sample_size}... and min_cluster_size={min_size}...", end=" ")
    
        # Fit HDBSCAN
        hdbscan = HDBSCAN(min_samples=sample_size, min_cluster_size=min_size, n_jobs=-1)
        clusters = hdbscan.fit_predict(sample_set1)
    
        # Calculate metrics
        n_clusters = len(set(clusters)) - (1 if -1 in clusters else 0)
        n_noise = list(clusters).count(-1)
        noise_pct = (n_noise / len(clusters)) * 100
    
        # Calculate ARI and NMI
        if n_clusters > 0:
            ari = adjusted_rand_score(labels_sample, clusters)
            nmi = normalized_mutual_info_score(labels_sample, clusters)
        
            # Silhouette score (only for non-noise points)
            if n_noise < len(clusters):
                mask = clusters != -1
                if mask.sum() > 1 and len(set(clusters[mask])) > 1:
                    silhouette = silhouette_score(sample_set1[mask], clusters[mask])
                else:
                    silhouette = np.nan
            else:
                silhouette = np.nan
        else:
            ari = np.nan
            nmi = np.nan
            silhouette = np.nan
    
        results.append({
            'min_sample_size': sample_size,
            'min_cluster_size': min_size,
            'n_clusters': n_clusters,
            'noise_count': n_noise,
            'noise_pct': noise_pct,
            'ari': ari,
            'nmi': nmi,
            'silhouette': silhouette
        })
    
        print(f"\n ✓ Clusters: {n_clusters}, Noise: {noise_pct:.1f}%, ARI: {ari:.3f}, NMI: {nmi:.3f}")
        count = count + 1

print("\n✓ Grid search complete!")
```

```python
# Convert to DataFrame
results_df = pd.DataFrame(results)
results_df = results_df.sort_values('ari', ascending=False)

print(f"\nTop 20 configurations by ARI:")
print(results_df.head(20).to_string(index=False))

print(f"\nTop 20 configurations by NMI:")
print(results_df.sort_values('nmi', ascending=False).head(20).to_string(index=False))
```

```python
fig, axes = plt.subplots(2, 3, figsize=(14, 10))
fig.suptitle("Clustering Results for Feature Set 1 (Spatial + Hourly)", fontsize=16, fontweight='bold')

# min_sample_size vs ARI
axes[0, 0].scatter(results_df['min_sample_size'], results_df['ari'], alpha=0.6)
axes[0, 0].set_xlabel('min_sample_size')
axes[0, 0].set_ylabel('ARI')
axes[0, 0].set_title('min_sample_size vs ARI')

# min_cluster_size vs ARI
axes[0, 1].scatter(results_df['min_cluster_size'], results_df['ari'], alpha=0.6)
axes[0, 1].set_xlabel('min_cluster_size')
axes[0, 1].set_ylabel('ARI')
axes[0, 1].set_title('min_cluster_size vs ARI')

# No. of Clusters vs ARI
axes[0, 2].scatter(results_df['n_clusters'], results_df['ari'], alpha=0.6)
axes[0, 2].set_xlabel('No. of Clusters')
axes[0, 2].set_ylabel('ARI')
axes[0, 2].set_title('No. of Clusters vs ARI')

# min_sample_size vs NMI
axes[1, 0].scatter(results_df['min_sample_size'], results_df['nmi'], alpha=0.6)
axes[1, 0].set_xlabel('min_sample_size')
axes[1, 0].set_ylabel('NMI')
axes[1, 0].set_title('min_sample_size vs NMI')

# min_cluster_size vs NMI
axes[1, 1].scatter(results_df['min_cluster_size'], results_df['nmi'], alpha=0.6)
axes[1, 1].set_xlabel('min_cluster_size')
axes[1, 1].set_ylabel('NMI')
axes[1, 1].set_title('min_cluster_size vs NMI')

# No. of Clusters vs NMI
axes[1, 2].scatter(results_df['n_clusters'], results_df['nmi'], alpha=0.6)
axes[1, 2].set_xlabel('No. of Clusters')
axes[1, 2].set_ylabel('NMI')
axes[1, 2].set_title('No. of Clusters vs NMI')

plt.tight_layout()
plt.show()
```

From the above results, the best parameters for feature set 1 would be sample size 50 and cluster size 200 to give the highest ARI and NMI.
However, for an ideal balance of minimal noise and moderate number of clusters, the most ideal parameter would be sample size 10 and cluster size 500.
In general, a smaller min sample and min cluster size would result in a higher ARI and NMI.

Applying HDBSCAN to a larger dataset of 500,000 using the best parameters scaled by a factor of 20 to account for the increased sample size

```python
# Randomly sample 200000 rows from the full dataset
FULL_SIZE = 500_000
print(f"Randomly sampling {FULL_SIZE:,} rows from {len(twitter):,} total rows...")

# Create random sample indices
np.random.seed(1)  # For reproducibility
sample_indices_full1 = np.random.choice(len(twitter), FULL_SIZE, replace=False)

# Get sampled features and labels
features_full1 = features_set1_scaled[sample_indices_full1]
labels_full1 = twitter['timezone'].iloc[sample_indices_full1]

print(f"✓ Random sample created: {len(features_full1):,} points\n")

# Get best configuration from grid search
best_min_sample_size = 200
best_min_cluster_size = 10000

print(f"Applying best configuration to 500,000 random sample:")
print(f"  best_min_sample_size: {best_min_sample_size}")
print(f"  best_min_cluster_size: {best_min_cluster_size}")

# Apply HDBSCAN to 500,000 sample with progress bar
print("\nClustering 500,000 points...")
import time
start_time = time.time()

hdbscan_full1 = HDBSCAN(min_samples=best_min_sample_size, min_cluster_size=best_min_cluster_size, n_jobs=-1)
clusters_full1 = hdbscan_full1.fit_predict(features_full1)


# Calculate metrics
n_clusters_full1 = len(set(clusters_full1)) - (1 if -1 in clusters_full1 else 0)
n_noise_full1 = list(clusters_full1).count(-1)
noise_pct_full1 = (n_noise_full1 / len(clusters_full1)) * 100

print(f"\nFinal Random Sample Results:")
print(f"  Clusters found: {n_clusters_full1}")
print(f"  Noise points: {n_noise_full1:,} ({noise_pct_full1:.2f}%)")

# Evaluate with progress
print("\nCalculating evaluation metrics...")
with tqdm(total=2, desc="Evaluation") as pbar:
    ari_full1 = adjusted_rand_score(labels_full1, clusters_full1)
    pbar.update(1)
    nmi_full1 = normalized_mutual_info_score(labels_full1, clusters_full1)
    pbar.update(1)

print(f"\nEvaluation Metrics:")
print(f"  ARI: {ari_full1:.3f}")
print(f"  NMI: {nmi_full1:.3f}")

# Add results to a subset dataframe
print("\nStoring results...")
twitter_full1 = twitter.iloc[sample_indices_full1].copy()
twitter_full1['cluster'] = clusters_full1

print(f"✓ Results stored in 'twitter_final' dataframe")

# Show cluster distribution
unique, counts = np.unique(clusters_full1, return_counts=True)
cluster_dist = pd.DataFrame({'cluster': unique, 'count': counts})
cluster_dist['percentage'] = (cluster_dist['count'] / len(clusters_full1) * 100)
cluster_dist = cluster_dist.sort_values('count', ascending=False)

print(f"\nTop 10 Largest Clusters:")
print(cluster_dist.head(10).to_string(index=False))

print(f"\n✓ Total processing time: {time.time() - start_time:.1f} seconds")
```
